# Import libraries and read files

```python
import pandas as pd
import numpy as np
import json 
import itertools
import seaborn as sns
cm = sns.light_palette("green", as_cmap=True)
def color_zero_red(val):
    color = 'red' if val == 0 else 'black'
    return 'color: %s' % color

# csv파일 read - video_data(video_id, publish_time, trending_date, views .etc)
videos_us = pd.read_csv('./youtube-new/USvideos.csv')
videos_ca= pd.read_csv('./youtube-new/CAvideos.csv')
videos_de = pd.read_csv('./youtube-new/DEvideos.csv')
videos_fr = pd.read_csv('./youtube-new/FRvideos.csv')
videos_gb = pd.read_csv('./youtube-new/GBvideos.csv')

# json파일 read - category_id, category_title
with open('./youtube-new/US_category_id.json') as f:
    categorys_us = json.load(f)
with open('./youtube-new/CA_category_id.json') as f:
    categorys_ca = json.load(f)
with open('./youtube-new/DE_category_id.json') as f:
    categorys_de = json.load(f)
with open('./youtube-new/FR_category_id.json') as f:
    categorys_fr = json.load(f)
with open('./youtube-new/GB_category_id.json') as f:
    categorys_gb = json.load(f)
```

```python
def saveCategoryTitles(f, datas, nation):
    n = len(f)
    for item in f:
        if int(item['id']) in datas.index:
            datas.loc[int(item['id'])][('title_'+nation)] = item['snippet']['title']
```

# 신뢰도 측정 모델 만들기

```python
reliability_table_1 = pd.DataFrame([], index=np.array(range(1,19)))
for i in range (1,19):
    reliability_table_1[i] = 100-(abs(reliability_table_1.index-i)/17)*100
style_reliability_table_1 = reliability_table_1.style.background_gradient(cmap=cm)
style_reliability_table_1
```

```python
# 등수의 순열
def worst_1(n):
    print('신뢰도 측정방법 1, n=', n)
    reliability = pd.DataFrame([], index=np.array(range(1,n+1)))
    for i in range (1,n+1):
        reliability[i] = 100-(abs(reliability.index-i)/(n-1))*100
    print(reliability)
    
    original=np.array(range(1,n+1))
    new_permutations = itertools.permutations(original)
    worst_list = []
    worst_mean = 100
    for new in new_permutations:
        sum_val=0
        for j in range (0,n):
            sum_val = sum_val + reliability[original[j]][new[j]]
        score = sum_val/n
        if score <= worst_mean:
            worst_mean = score
            worst_list=new
    print('최악의경우 reliability =',worst_mean)
    print(worst_list)
    print()
    
for i in range(5,11):
    worst_1(i)
```

```python
# 신뢰도 측정 방법 1의 worst case : 역순 정렬된 배열
n=18
list_original_1=np.array(range(1,n+1)) # [1, 2, ..., 18]
print(list_original_1)
list_new_1 = np.array(range(n,0,-1)) # [18, 17, ..., 1]
print(list_new_1)
sum_1 = 0
for i in range (0,n):
    sum_1 = sum_1 + reliability_table_1[list_original_1[i]][list_new_1[i]]
mean_1 = sum_1/n
mean_1
```

```python
n=18
reliability_table_2 = pd.DataFrame([], index=np.array(range(1,n+1)))
for i in range (1,n+1):
    if n-i > i-1:
        max_val = n-i
    else:
        max_val = i-1
    reliability_table_2[i] = 100-(abs(reliability_table_2.index-i)/max_val)*100
style_reliability_table_2 = reliability_table_2.style.background_gradient(cmap=cm)
style_reliability_table_2
```

```python
def worst_2(n):
    print('신뢰도 측정방법 2, n=', n)
    reliability = pd.DataFrame([], index=np.array(range(1,n+1)))
    for i in range (1,n+1):
        if n-i>i-1:
            max_val = n-i
        else:
            max_val = i-1
        reliability[i] = 100-(abs(reliability.index-i)/max_val)*100
    print(reliability)
    
    original=np.array(range(1,n+1))
    new_permutations = itertools.permutations(original)
    worst_list = []
    worst_mean = 100
    for new in new_permutations:
        sum_val=0
        for j in range (0,n):
            sum_val = sum_val + reliability[original[j]][new[j]]
        score = sum_val/n
        if score <= worst_mean:
            worst_mean = score
            worst_list=new
    print('최악의경우 reliability =',worst_mean)
    print(worst_list)
    print()
    
for i in range(5,11):
    worst_2(i)
```

```python
# 신뢰도 측정 방법 2의 worst case (9씩 차이나는 수)
n=18
list_original_2=np.array(range(1,n+1)) # [1, 2, ..., 18]
print(list_original_2)
list_new_2 =  (list_original_2+8)%18 + 1 # [10, 11, ..., 18, 1, 2, ..., 8, 9]
print(list_new_2)
sum_2 = 0
for i in range (0,n):
    sum_2 = sum_2 + reliability_table_2[list_original_2[i]][list_new_2[i]]
mean_2 = sum_2/n
mean_2
```

```python
n=18
reliability_table_3 = pd.DataFrame(np.eye(n)*100, index=np.array(range(1,n+1)), columns=np.array(range(1, n+1)))
style_reliability_table_3 = reliability_table_3.style.background_gradient(cmap=cm)
style_reliability_table_3
```

```python
def worst_3(n):
    print('신뢰도 측정방법 3, n=', n)
    reliability = pd.DataFrame(np.eye(n)*100, index=np.array(range(1,n+1)), columns=np.array(range(1,n+1)))
    print(reliability)
    
    original=np.array(range(1,n+1))
    new_permutations = itertools.permutations(original)
    worst_list = []
    worst_mean = 100
    for new in new_permutations:
#         sum_val=0
#         for j in range (0,n):
#             sum_val = sum_val + reliability[original[j]][new[j]]
#         score = sum_val/n
        score = (sum(original==new)/n)*100
        if score <= worst_mean:
            worst_mean = score
            worst_list=new
    print('최악의경우 reliability =',worst_mean)
    print(worst_list)
    print()
    # 전체 신뢰도는 (sum(list1==list2)/n)*100으로 계산할 수 있음
    
for i in range(5,11):
    worst_3(i)
```

```python
# 신뢰도 측정 방법 3의 worst case
n=18
list_original_3=np.array(range(1,n+1)) # [1, 2, ..., 18]
print(list_original_3)
list_new_3 = (list_original_3+8)%18 + 1 # [10, 11, ..., 18, 1, 2, ..., 8, 9]
print(list_new_3)
mean_2 = (sum(list_original_3==list_new_3)/n)*100
mean_2
```

```python
def showResult(result_val):
    result = category_datas.copy()
    result['English'] = kedoo_rank['English']

    result['result'] = result_val
    reliability_1 = []
    reliability_2 = []
    reliability_3 = []
    for i in result.index:
        reliability_1.append(reliability_table_1[result['English'][i]][result['result'][i]])
        reliability_2.append(reliability_table_2[result['English'][i]][result['result'][i]])
        reliability_3.append(reliability_table_3[result['English'][i]][result['result'][i]])
    result['diffrence'] = result['English']-result['result']
    result['reliability_1'] = reliability_1
    result['reliability_2'] = reliability_2
    result['reliability_3'] = reliability_3
    
    print('원본',result['English'].values)
    print()

    print('결과',result['result'].values)
    print('차이',(result['English']-result['result']).values) # 음수:하락, 양수:상승
    print('신뢰도 1 : ', np.mean(result['reliability_1']))
    print('신뢰도 2 : ', np.mean(result['reliability_2']))
    print('신뢰도 3 : ', np.mean(result['reliability_3']))
    print()

    return result
```

```python
result_best=np.array(kedoo_rank['English'])
result_best_table = showResult(result_best)
result_best_table
```
