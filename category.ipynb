```python
import pandas as pd
```

```python
videos_us = pd.read_csv('./youtube-new/USvideos.csv')
videos_ca= pd.read_csv('./youtube-new/CAvideos.csv')
videos_de = pd.read_csv('./youtube-new/DEvideos.csv')
videos_fr = pd.read_csv('./youtube-new/FRvideos.csv')
videos_gb = pd.read_csv('./youtube-new/GBvideos.csv')
```

```python
# videos파일 속성 출력하기
videos_us.columns
#videos_ca.columns
# videos_de.columns
# videos_fr.columns
# videos_gb.columns
```

```python
import numpy as np
used_category_us = list(np.unique(videos_us.category_id))
used_category_ca = list(np.unique(videos_ca.category_id))
used_category_fr = list(np.unique(videos_fr.category_id))
used_category_gb = list(np.unique(videos_gb.category_id))
used_category_de = list(np.unique(videos_de.category_id))

used_category_us
```

```python
import json 
with open('./youtube-new/US_category_id.json') as f:
    category_data_us = json.load(f)
with open('./youtube-new/CA_category_id.json') as f:
    category_data_ca = json.load(f)
with open('./youtube-new/DE_category_id.json') as f:
    category_data_de = json.load(f)
with open('./youtube-new/FR_category_id.json') as f:
    category_data_fr = json.load(f)
with open('./youtube-new/GB_category_id.json') as f:
    category_data_gb = json.load(f)
```

```python
categorys_us = category_data_us['items']
categorys_ca = category_data_ca['items']
categorys_fr = category_data_fr['items']
categorys_gb = category_data_gb['items']
categorys_de = category_data_de['items']
# for item in categorys_us:
#     print(item['id'] + ' '+ item['snippet']['title'])
```

```python
# len(used_category_us) # 16
# len(used_category_ca) # 17
# len(used_category_fr) # 18
# len(used_category_gb) # 16
len(used_category_de) # 18
```

```python
# category 합치기
used_category_1 = np.union1d(videos_us.category_id, videos_ca.category_id)
used_category_2 = np.union1d(videos_de.category_id, videos_fr.category_id)
used_category_2 = np.union1d(used_category_2, videos_gb.category_id)
used_category = np.union1d(used_category_1, used_category_2)
used_category # (us + ca) + ((de + fr) + gb)
```

```python
used_categorys_nation = pd.DataFrame([],index=['US','CA','FR','GB','DE'])
for item in used_category:
    isIn=[]
    if item in used_category_us:
        isIn.append(1)
    else:
        isIn.append(0)
    if item in used_category_ca:
        isIn.append(1)
    else:
        isIn.append(0)
    if item in used_category_fr:
        isIn.append(1)
    else:
        isIn.append(0)
    if item in used_category_gb:
        isIn.append(1)
    else:
        isIn.append(0)
    if item in used_category_de:
        isIn.append(1)
    else:
        isIn.append(0)
    used_categorys_nation[item] = isIn
used_categorys_nation
```

```python
i=0
title=[]
for item in categorys_us:
    if int(item['id']) == used_category[i]:
        title.append(item['snippet']['title'])
        i = i + 1

used_categorys_nation.loc['title'] = title
used_categorys_nation
```
